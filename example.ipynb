```python
import math
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
from sklearn import datasets
%matplotlib inline
matplotlib.rc('font', family='Arial Unicode MS',)  ## 解决图上文字乱码的问题
```

```python
from tda import TDA
```

```python
iris = datasets.load_iris()
data = iris.data
y = iris.target
```

```python
def distance(a, b):
    s = sum( (i-j) **2 for i, j in zip(a, b))
    return math.sqrt(s)
```

```python
def L_inf(n, i):
    L = 0
    for j in range(len(n)):
        dist = distance(n[i], n[j])
        if dist > L:
            L = dist
    return L
```

```python
t = TDA(distance, [(L_inf, 25, 0.8)], 6)
t.fit(data)
t.binums, len(t.clusters)
```

```python
t.dye(lambda d, i: y[i], figsize=10, Pk=1)
```
